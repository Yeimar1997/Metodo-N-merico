### Clculo de promedios y de valores medios

**La media:** Se define como la suma de N elementos dividida entre N. Se trata una medida bastante conocida entre la gente, aunque tiene el inconveniente de que es muy susceptible a valores extremos.

***Media: mean(data)*** 

La funcin **mean()** devuelve la media aritmtica de los datos de una muestra o poblacin. Esta media se calcula sumando todos los valores y despus el resultado de la suma se divide entre el nmero de elementos. 

**Algoritmo**

```python
import statistics as stats

edades = [46, 54, 42, 46, 32, 30]
print(stats.mean(edades)) 
```

**La mediana:** Es el valor que dentro del conjunto de datos es menor que el 50% de los datos y mayor que el 50% restante.

***Mediana: median(data)***

La funcin **median()** devuelve la mediana o valor medio de los datos. La mediana es el valor central de un grupo de nmeros ordenados por tamao. Si la cantidad de elementos es par, la mediana es el promedio de los dos nmeros centrales.

**Algoritmo**


```python
Dias = [22, 23, 26, 26, 26, 34, 34, 38]
print(stats.median(Dias))  
```

**La moda:** Es el valor ms repetido (solo aplicable a variables discretas).

***Moda: mode(data)***

La funcin **mode()** devuelve el valor ms frecuente de los datos.

**Algoritmo**

```python
edades = [22, 23, 26, 26, 26, 34, 34, 38, 40, 41, 38, 38, 38, 22, 12, 23, 22, 25]
print(stats.mode(edades))  
```

# Medidas de dispersin

**varianza:**La varianza es la media de las diferencias con la media elevadas al cuadrado.
 
***Varianza (poblacin): pvariance(data, mu=None)***

La funcin **pvariance()** devuelve la varianza de una poblacin de datos. Esta funcin es equivalente a variance() pero se aplica a una poblacin en lugar de a una muestra. La funcin pvariance() para obtener la varianza primero calcula la media de los elementos:

edades = [22, 23, 26, 26, 26, 34, 34, 38, 40, 41]
media = (22 + 23 + 26 + 26 + 26 + 34 + 34 + 38 + 40 + 41) / 10 = 31

A cada nmero se le resta la media obtenida y se eleva al cuadrado. Despus, se suman todos los resultados:

dif_cuadrado = (22-31)**2 + (23-31)**2 + (26-31)**2 + (26-31)**2 + (26-31)**2 + (34-31)**2 + (34-31)**2 + (38-31)**2 + (40-31)**2 + (41-31)**2 = 468

Finalmente, se calcula la media de todas las diferencias al cuadrado:

varianza = 468 / 10 = 46.8

**Algoritmo**

```python
edades = [22, 23, 26, 26, 36, 34, 34, 38, 40, 41]
print(stats.pvariance(edades))
```

***Varianza (muestra): variance(data, xbar=None)***

La funcin variance() devuelve la varianza de una muestra de datos. La varianza mide en qu medida se dispersan los valores de una muestra alrededor de un valor central.

El argumento xbar es opcional y si se facilita debe ser la media de los datos. En caso contrario, la funcin calcular este dato automticamente.

Para calcular la varianza de una poblacin entera utilizar la funcin pvariance().

**Algoritmo**


```python
muestra_edades = [22, 35, 26, 12, 34, 38, 42]
print(stats.variance(muestra_edades))  
```

# Tipos de Tipos de grficos 

Un  grfico  estadstico  es  una  representacin  visual  de  una  serie  de  datos  estadsticos. 

**grafico de barra**

* Este grafico representa  los datos  cualitativos de una distribucin

* Un grfico de barras es una representacin grfica en un eje cartesiano de las frecuencias de una variable cualitativa o discreta.

**Algortimo de un ejemplo**

**Ejemplo** 

Tenemos una muestra de 50  ventas de refresco realizar el diagrama de barra y el diagrama de torta 

**Algoritmo**

```python
import numpy as np
import matplotlib.pyplot as plt

bebidas = ['Coke classic', 'Died coke', 'Dr pepper', 'Pepsi', 'sprite'] #Nombres de las bebidas 
Cantidad = [19, 8, 5, 13, 5] # Nro Cantidades

plt.bar(range(5), Cantidad, edgecolor='black')

plt.xticks(range(5), bebidas, rotation=60)
plt.title("REFRESCOS")
plt.ylim(min(Cantidad)-1, max(Cantidad)+1) #Muestra la cantidades en la grafica 
plt.show()
```

**Diagrama de torta**

* Esta grafica representa  datos  resumidos  mediante una distribucion de frecuencia.

* Un grfico de pastel es un grfico circular, el cual est dividido de tal manera que se asemeja a las tajadas de un pastel. 

* Este grfico divide tus datos en sectores, ilustrando cada proporcin numrica con el fin de mostrar la composicin como un todo.


```python
cantidad = [19, 8, 5, 13, 5]
bebidas = ['Coke classic', 'Died coke', 'Dr pepper', 'Pepsi', 'sprite']

explode = [0.38, 0.16, 0.1, 0.26, 0.1]  #Sirve para separar quesitos del centro de la tarta

plt.pie(cantidad, labels=bebidas, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title('REFRESCOS')
plt.show()
```

***Diagramas de caja y bigotes***

* Un diagrama de cajas y bigotes es una manera conveniente de mostrar visualmente grupos de datos numricos a travs de sus cuartiles.

Las lneas que se extienden paralelas a las cajas se conocen como bigotes, y se usan para indicar variabilidad fuera de los cuartiles superior e inferior. Los valores atpicos se representan a veces como puntos individuales que estn en lnea con los bigotes. Los diagramas de cajas y bigotes se pueden dibujar vertical u horizontalmente.

los grficos de cajas y bigotes son una excelente forma de examinar rpidamente uno o ms conjuntos de datos grficamente. Aunque parezcan primitivos en comparacin con un Histograma o un Grfico de Densidad, tienen la ventaja de ocupar menos espacio, lo cual es til cuando se comparan distribuciones entre muchos grupos o conjuntos de datos. 


**Algoritmo**

```python
import matplotlib.pyplot as plt
import numpy as np

X1 = np.random.normal(0, 1, 500)
X2 = np.random.normal(0.3, 1, 500)

# El diagrama de caja simple
plt.boxplot(X1)
plt.show()

# Cambiar caractersticas
plt.boxplot(X1, notch=True, sym="o") # Use sym="" to shown no fliers; also showfliers=False
plt.show()

# Mostrar mltiples diagramas de caja en la misma ventana
plt.boxplot((X1, X2), notch=True, sym="o", labels=["Set 1", "Set 2"])
plt.show()

# Personalizacin avanzada de la grfica de caja
line_props = dict(color="r", alpha=0.3)
bbox_props = dict(color="g", alpha=0.9, linestyle="dashdot")
flier_props = dict(marker="o", markersize=17)
plt.boxplot(X1, notch=True, whiskerprops=line_props, boxprops=bbox_props, flierprops=flier_props)
plt.show()
```

# Graficando distribuciones

**Histogramas**

Es una de las mejores maneras de describir una variable es representar los valores que aparecen en el conjunto de datos y el nmero de veces que aparece cada valor.

En Python, podemos graficar fcilmente un histograma con la ayuda de la funcin hist de matplotlib, simplemente debemos pasarle los datos y la cantidad de contenedores en los que queremos dividirlos. Por ejemplo, podramos graficar el histograma de una distribucin normal del siguiente modo.

### Algotimo


```python
# importando modulos necesarios
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np 
from scipy import stats 
import seaborn as sns 

np.random.seed(2016) # replicar random

# parametros esteticos de seaborn
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (8, 4)})
# Graficando histograma
mu, sigma = 0, 0.2 # media y desvio estandar
datos = np.random.normal(mu, sigma, 1000) #Creacion de muestra de datos

# Histograma de distribucin normal.
cuenta, cajas, ignorar = plt.hist(datos, 20)
plt.ylabel('Frequencia')
plt.xlabel('Valores')
plt.title('Histograma')
plt.show()
```

## Distribuciones Discretas

Las distribuciones discretas son aquellas en las que la variable puede tomar solo algunos valores determinados. Los principales exponentes de este grupo son las siguientes:

***Distribucin Poisson***

La Distribucin Poisson esta dada por la formula: En dnde r es un entero (r0) y  es un nmero real positivo.


En Python la podemos generar fcilmente con la ayuda de **scipy.stats**, paquete que utilizaremos para representar a todas las restantes distribuciones a lo largo de todo el artculo.

### Algortimos



```python
# Graficando Poisson
mu =  3.6 # parametro de forma 
poisson = stats.poisson(mu) # Distribucin
x = np.arange(poisson.ppf(0.01),
              poisson.ppf(0.99))
fmp = poisson.pmf(x) # Funcin de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribucin Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

# histograma
aleatorios = poisson.rvs(1000)  # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('Frequencia')
plt.xlabel('Valores')
plt.title('Histograma Poisson')
plt.show()



```

## Distribucin Binomial

La Distribucin Binomial esta dada por la formula: En dnde r con la condicin 0rN y el parmetro N (N>0) son enteros; y el parmetro p (0p1) es un nmero real.




### Algotimo



```python
# Graficando Binomial
N, p = 30, 0.4 # parametros de forma 
binomial = stats.binom(N, p) # Distribucin
x = np.arange(binomial.ppf(0.01),
              binomial.ppf(0.99))
fmp = binomial.pmf(x) # Funcin de Masa de Probabilidad
plt.plot(x, fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribucin Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()

# Histograma
aleatorios = binomial.rvs(1000)  # genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('Frequencia')
plt.xlabel('Valores')
plt.title('Histograma Binomial')
plt.show()
```
